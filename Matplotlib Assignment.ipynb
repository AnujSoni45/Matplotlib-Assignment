Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

# Answer
Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in a wide range of formats. It provides a flexible and comprehensive set of tools for creating various types of plots and charts, making it a go-to choice for data visualization in scientific research, data analysis, and other fields. Matplotlib was originally developed by John D. Hunter in 2003 and has since gained widespread adoption in the Python community.

Matplotlib is used for a variety of purposes, including:

Data Visualization: Matplotlib allows you to create a wide range of plots and charts, making it easy to visualize data and gain insights from it.

Report Generation: It's commonly used in scientific research and data analysis to generate high-quality plots and charts for inclusion in reports and publications.

Exploration and Analysis: Data analysts and scientists often use Matplotlib to explore and analyze data by creating visual representations of datasets.

Teaching and Learning: Matplotlib is a valuable tool for teaching and learning data visualization techniques and concepts in Python.

Dashboard Creation: Matplotlib can be used to create static or interactive dashboards for displaying data in a user-friendly way.


(1) Line Plot: Line plots are used to display data points connected by lines. They are commonly used to visualize trends and patterns in data over a continuous range, such as time.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

(2) Bar Plot: Bar plots are used to represent categorical data with rectangular bars. They are suitable for comparing data across different categories.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 12, 25, 15, 30]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

(3) Scatter Plot: Scatter plots are used to visualize the relationship between two variables. Each data point is represented as a point on the graph.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

(4) Histogram: Histograms are used to represent the distribution of a continuous dataset. They divide the data into bins and display the frequency of data points within each bin.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  
plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

(5) Pie Chart: Pie charts are used to display the composition of a categorical dataset as a circular chart divided into slices, where each slice represents a category's proportion.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [30, 20, 25, 15]

plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

# Answer
A scatter plot is a data visualization technique used to display the relationship or distribution between two continuous variables. In a scatter plot, each data point is represented as a point on a two-dimensional plane, with one variable plotted on the x-axis and the other on the y-axis. Scatter plots are helpful for visualizing patterns, clusters, trends, and correlations between two variables.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, label='Data Points', color='blue', marker='o')
plt.xlabel('X-axis')     
plt.ylabel('Y-axis')     
plt.title('Scatter Plot')  
plt.legend()               
plt.grid(True)          

plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
# Answer
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Create a figure with 2 rows and 2 columns
plt.figure(figsize=(10, 8))  

# Line Plot 1
plt.subplot(2, 2, 1)  
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1, label='Line 1', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 1')

# Line Plot 2
plt.subplot(2, 2, 2)  
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2, label='Line 2', color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 2')

# Line Plot 3
plt.subplot(2, 2, 3)  
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3, label='Line 3', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 3')

# Line Plot 4
plt.subplot(2, 2, 4)  
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4, label='Line 4', color='purple')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Answer
A bar plot, also known as a bar chart or bar graph, is a data visualization technique used to represent categorical data with rectangular bars. Each bar's length or height is proportional to the value it represents. Bar plots are useful for comparing and displaying the relationship between different categories or groups of data. They are commonly used to show discrete, non-continuous data and make it easy to visualize comparisons, trends, or rankings.

Bar plots are used for various purposes, including:

Comparisons: You can use bar plots to compare quantities or values across different categories or groups. For example, comparing the profits of different companies.

Ranking: Bar plots can be used to rank categories based on a particular metric, making it clear which category has the highest or lowest value.

Frequency Distribution: Bar plots are useful for displaying the frequency distribution of categorical data.

Part-to-Whole Relationships: Stacked bar plots or grouped bar plots can represent part-to-whole relationships within categories.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(8, 6))  
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Company Profits (Vertical Bar Plot)')
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))  
plt.barh(company, profit, color='lightcoral')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest profit at the top
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Answer
A box plot, also known as a box-and-whisker plot, is a data visualization technique used to display the distribution and summary statistics of a dataset. It provides a visual summary of the central tendency, spread, and potential outliers in the data. Box plots are particularly useful for:

Identifying Central Tendency: You can see the median (the middle value), which represents the central tendency of the data.

Understanding Spread: The interquartile range (IQR) is displayed as a box, and it shows the spread of the middle 50% of the data.

Detecting Skewness: Box plots can reveal if the data is skewed or symmetric.

Detecting Outliers: Outliers are displayed as individual points beyond the "whiskers" of the plot.

Comparing Distributions: You can use multiple box plots side by side to compare the distributions of different datasets.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(8, 6))  
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], vert=False)  # vert=False for horizontal plot
plt.xlabel('Value')
plt.ylabel('Box')
plt.title('Box Plot of Box 1 and Box 2')
plt.show()
```
